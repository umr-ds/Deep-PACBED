# Experimental PACBED test dataset

The **Exp** dataset consists of a collection of 1030 PACBED images of four materials (Si, GaP, GaAs, AlAs) from different experimental measurements.

For each material, the `img` folders contain the images, the `md_split` folders contain the metadata (annotations) for each image, split into train and test subsets, while the combined metadata is provided in the `full.csv` files.

### Imports

```python
import glob
import pandas as pd
```

### Check for completeness

```python
imgs = glob.glob("Exp/*/img/**/*.png", recursive=True)
assert len(imgs) == 1030, f"{1030 - len(imgs)} missing image files"
```

```python
mds = glob.glob("Exp/**/*.csv", recursive=True)
assert len(mds) == 12, f"{12 - len(mds)} missing md files"
```

```python
meta_count = 0
for md in mds:
    meta = pd.read_csv(md)
    meta_count += len(meta)
assert meta_count == 2060, f"{2060 - meta_count} missing md entries"
```
